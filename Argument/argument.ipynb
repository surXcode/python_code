* It define how fast we can access and modify the variable.

# Python follow LEGB rule:

* LEGB rule: Local enclose globle built in 

   L:Local
   E: enclosed/ Non-local
   G: globle
   B: built in 

      # L < E < G < B

#  Global space variable: 

* If we declare any variable in main space then it is known as global space variable.

  * we  can access global variable value at main space as well as inside the functional space also.

  * we can not modify the global scope variable inside functional space, in order do that we have to use the global keyword.

  # Local space variable:

   If we declare a variable inside of a function known as local space variable 

   we can not access the local scope of variable at main space, it can we access inside of the function.

#  Enclosed/Non-local:

* if we declare inside the nested function then it is consider as enclosed and non local space of variable.

* we can access this enclosed variavble at local and non local variable.

# Non global keyword:

* if we try to modify the non local variable or enclosed varibale inside the local space by default its not possible, in order do that we have to use non local keyword.

# Built in scope of varible:
